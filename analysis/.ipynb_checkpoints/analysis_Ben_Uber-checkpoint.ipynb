```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# (Optional) for inline plots in Jupyter
%matplotlib inline
```

```python
# Adjust path if needed
driver_trips = pd.read_csv("Driver/driver_lifetime_trips-0.csv")

# Convert "Local Dropoff Timestamp" into an actual datetime
driver_trips["Local Dropoff Timestamp"] = pd.to_datetime(
    driver_trips["Local Dropoff Timestamp"],
    errors="coerce"
)

# Make day_of_week, hour_of_day columns
driver_trips["day_of_week"] = driver_trips["Local Dropoff Timestamp"].dt.day_name()
driver_trips["hour_of_day"] = driver_trips["Local Dropoff Timestamp"].dt.hour
```

```python
print(driver_trips.head())
print(driver_trips.info())
```

```python
avg_dist_by_day = driver_trips.groupby("day_of_week")["Trip Distance (miles)"].mean()
avg_dist_by_day.plot(kind="bar", figsize=(8, 4), color="skyblue")
plt.title("Average Trip Distance by Day of the Week")
plt.ylabel("Miles")
plt.show()
```

```python
# Create a pivot: rows=day_of_week, cols=hour_of_day, values=Trip Distance (miles)
dist_pivot = driver_trips.pivot_table(
    index="day_of_week",
    columns="hour_of_day",
    values="Trip Distance (miles)",
    aggfunc="mean"
)

plt.figure(figsize=(10, 6))
sns.heatmap(dist_pivot, cmap="Blues", annot=True, fmt=".1f")
plt.title("Avg Trip Distance by Day & Hour")
plt.ylabel("Day of Week")
plt.xlabel("Hour of Day")
plt.show()
```

```python
driver_payments = pd.read_csv("Driver/driver_payments-0.csv")
driver_payments["Local Timestamp"] = pd.to_datetime(
    driver_payments["Local Timestamp"],
    errors="coerce"
)
driver_payments["day_of_week"] = driver_payments["Local Timestamp"].dt.day_name()
driver_payments["hour_of_day"] = driver_payments["Local Timestamp"].dt.hour
```

```python
pay_sums = driver_payments.pivot_table(
    index="day_of_week",
    columns="hour_of_day",
    values="Local Amount",
    aggfunc="sum"
)

plt.figure(figsize=(10,6))
sns.heatmap(pay_sums, cmap="Greens", annot=True, fmt=".2f")
plt.title("Total Payment Amount by Day & Hour")
plt.show()
```
